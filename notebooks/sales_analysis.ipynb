```python
monthly_sales = (
    df
    .set_index("Order Date")
    .resample("M")["Sales"]
    .sum()
    .reset_index()
)

monthly_sales.head()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(monthly_sales["Order Date"], monthly_sales["Sales"])
plt.title("Monthly Sales Trend")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
```

The monthly sales trend shows recurring peaks toward the end of the year, indicating seasonality in retail demand. Mid-year periods appear relatively stable with occasional dips, suggesting potential low-demand windows for targeted promotions.

```python
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet
```

```python
df = pd.read_csv("../data/superstore.csv", encoding="latin1")
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
```

```python
monthly_sales = (
    df
    .set_index("Order Date")
    .resample("M")["Sales"]
    .sum()
    .reset_index()
)
```

```python
monthly_sales.head()
```

```python
forecast_df = monthly_sales.rename(
    columns={"Order Date": "ds", "Sales": "y"}
)
```

```python
model = Prophet()
model.fit(forecast_df)
```

```python
future = model.make_future_dataframe(periods=12, freq="M")
```

```python
forecast = model.predict(future)
```

```python
model.plot(forecast)
plt.show()
```

```python
forecast_output = forecast[["ds", "yhat", "yhat_lower", "yhat_upper"]]
forecast_output.to_csv("../outputs/sales_forecast.csv", index=False)
```
